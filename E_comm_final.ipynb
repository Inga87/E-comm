### Одинаковые `customer_id`, `order_id`, `product`,`quantity`

---

В данных есть наблюдения, когда все параметры повторяются кроме даты, как вариант, выгрузка включает в себя события, но не имеет параметр события, то есть присутствуют отдельные наблюдения, когда заказ был оплачен и когда был доставлен, например.

Такие события нужно удалить, так как для анализа нам не нужны события, а только факт покупки, особенно учитывая, что не все заказы дублируются, те выгрузка получилась не последовательной. Нужно отметить, что факт покупки не определен. Будем считать, что факт покупки наступает первым и все последующие события, связанные с заказом, являются дубликатом. 

Таким образом, удалим все наблюдения, которые одинаковые кроме даты и оставим только первые.

### Несколько customer_id для одного order_id
---
В данных обнаружено, что для одного и того же `order_id` могут быть разные `customer_id`, это выглядит некорректно, так как номер заказа должен быть уникальный. После проверки оказалось, что большая часть позиций в таких заказах абсолютно одинаковая и различия только в `customer_id` при одинаковом `order_id`, поэтому такие позиции были удалены, как дубликаты.


>**Вывод: Доля удаленных наблюдений от первоначального датасета 0.5%, эффект на выручку 0.5%**

*Также были обнаружены неточности по цене одного и того же товара, и различия в номенклатуре в одинаковом номере заказа с разным номером клиента.*

*Также было обнаружено, что для уникального `order_id` может быть несколько дат. Можно предположить, что речь идет о добавлении товара в заказ вручную уже после оплаты на сайте, так как в данном случае товары не совпадают.*

```python
#выберем все номера заказов, где на один заказ приходится 2 и более уникальных customer_id 

multiple_customer_id_for_same_order_id = df_v1.groupby(['order_id']).agg(
    {'customer_id':'nunique','revenue':'sum'}
).sort_values(by='customer_id',ascending = False) #группировка по order_id

multiple_customer_id_for_same_order_id = multiple_customer_id_for_same_order_id[
    multiple_customer_id_for_same_order_id['customer_id'] > 1].reset_index() #выборка только, где кол-во клиентов более 1

#список заказов, где кол-во клиентов более двух
multiple_customer_id_for_same_order_id_orderslist = multiple_customer_id_for_same_order_id['order_id'].to_list()
```

```python
print('Выручка по таким заказам составляет',multiple_customer_id_for_same_order_id['revenue'].sum())
print('Кол-во таких заказов составляет',multiple_customer_id_for_same_order_id.shape[0])
```

```python
#посмотрим детали таких заказов
(
    df_v1.query('order_id == @multiple_customer_id_for_same_order_id_orderslist')
    .pivot_table(index = ['order_id','customer_id','product'], values = ['quantity','revenue'], aggfunc = 'sum'))
```

```python
df_v1[df_v1['order_id'] == 69485]
```

Странная ситуация, похожа на баг при сборе данных, к одному и тому же заказу может быть привязано два клиента, причем дата совершенно разная, а товары и их количество *в большинстве случаев* совпадают. Нужно обязательно это обсудить с теми, кто ответственен за сбор данных, так как номер заказа не может повторяться, только если это опять не вариант с событиями, но почему присваивается другой клиент не понятно. 

Посмотрим далее, как и что вообще заказывают клиенты, которые попали в этот список, является ли для них это единственный заказ или у них много заказов. 

```python
# список клиентов, которые попадают в одни и те же заказы
multiple_customer_id_for_same_order_id_customerlist = df_v1.query('order_id == @multiple_customer_id_for_same_order_id_orderslist')['customer_id'].to_list()
```

Самый эффективный способ был бы взять уже существующие товарные категории в крупном онлайн магазине, например в Озоне, причем взять все уровни категорий, так как наш датасет это всего лишь случайная выборка, соответственно товаров может быть больше, чем есть в датасете. Для этого нужно спарсить динамический сайт (время не позволяет обучится новым способам парсинга). В связи с чем было принято решение сделать категоризацию только одного уровня, максимально сужая количество категорий, и сделать ее вручную.

Ограничения ручной классификации понятны, их вряд ли можно будет повторить на остальные товары, тогда как с парсингом можно было бы, поэтому парсинг конечно однозначно лучше.

Можно было бы использовать лемматизацию, но она бы не сильно упростила задачу, так как имеем дело не с текстом, а со списком товаров, которые в большинство случаев в именительном падеже. Также лемматизация происходит долго, даже если предварительно очистить дубликаты названий. 

Поэтому категоризация была сделана в несколько этапов:
1. Отсечение всех товаров, которые имеют отношение к цветам
2. Использование категорий Озона для оставшейся части (файл размещен в интернете)
3. Ручная проверка получившихся категорий и товаров в них, с целью выявления ошибок и сужения кол-ва категорий.

Гипотеза N. Параметр клиентов группы X отличается от параметра клиентов группы Y

H0 = Параметр клиентов группы X равен параметру клиентов группы Y

H1 = Параметр клиентов группы X не равен параметру клиентов группы Y

Сделаем поправку на множественную проверку гипотез, воспользуемся наиболее простой, поправкой Бонферрони, поделим альфа на кол-во экспериментов - 15.

*(для более понятной структуры и визуализации для каждого кластера проведем тест с остальными кластерам, то есть будут повторы с целью иметь под рукой полную информцию по кластеру внутри подраздела, а не повторной проверки)*

```python
# создадим функцию, которая будет рассчитывать стат значимость по критерию Манн-Уитни и сравнивать с критерием альфа 5%
def hypo(variable,label,label2,sample1, sample2):
    
    '''эта часть функции прописыват гипотезу 0 и 1'''
    
    print('\033[1m','Гипотеза. Параметр',variable,'клиентов группы',label,
          'отличается от',variable, 'клиентов группы',label2,'\033[0m')
    print('H0 = Параметр', variable, 'клиентов группы',label, 'равен',variable, 'клиентов группы',label2)
    print('H1 = Параметр', variable, 'клиентов группы',label, 'НЕ равен',variable, 'клиентов группы',label2)
    print('----------------------------------------------')
    
    '''эта часть функции выводит P-value'''
    print("P-value равно {0:.3f}".format(stats.mannwhitneyu(sample1, sample2, True, 'two-sided')[1]))
    
    '''эта часть функции рассчитывает стат значимость и выводит результат'''
    alpha = 0.05 / 15
    if (stats.mannwhitneyu(sample1, sample2)[1]) < alpha:
        print('\033[92m',"Отвергаем нулевую гипотезу",'\033[0m')
    else:
        print('\033[93m',"Не получилось отвергнуть нулевую гипотезу",'\033[0m') 
```
